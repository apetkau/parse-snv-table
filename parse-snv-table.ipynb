# Read input SNV table

Input file comes from <https://snvphyl.readthedocs.io/en/latest/examples/data/Listeria_monocytogenes_results.tar.gz>.

```r
table <- read.table('snvTable.tsv', comment.char='', header=TRUE)
table
```

# Select rows where SNVs differ

```r
col1 <- 'CFSAN002349'
col2 <- 'CFSAN023463'
table[table[col1] != table[col2],]
```

# Select table subset where columns differ

```r
col1 <- 'CFSAN002349'
col2 <- 'CFSAN023463'
table_subset <- table[table[col1] != table[col2],c('X.Chromosome', 'Position', 'Status', col1, col2)]
table_subset
```

# Save table subset to file

```r
write.table(table_subset, file='snvTable-subset.tsv', sep='\t')
```
